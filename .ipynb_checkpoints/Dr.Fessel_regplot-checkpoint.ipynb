# Seaborn: regplot
The overall purpose of regplot is to overlay your scatter points with a liner regression model. 
A linear regression model is looking for a relationship between X and Y.

Use `.sample(n= , random_state= )` to randomly pick samples for regression plot.

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
diamonds = sns.load_dataset('diamonds')

diamonds.shape
```

```python
diamonds.head()
```

```python
# Selecting 200 points randomly
# random_state initialize a random number generator

diamonds = diamonds.sample(n=200, random_state=44)

diamonds.shape
```

# Basics
Use `.regplot(x=' ', y=' ', data= )`

```python
sns.set_style('dark')
```

```python
sns.regplot(x='carat', y='price', data=diamonds)
```

```python
# Turning off a fit line, leaving a simple scatter plot

sns.regplot(x='carat', y='price', data=diamonds, 
            fit_reg=False
           )
```

```python
# Turning off points, leaving a linear plot

sns.regplot(x='carat', y='price', data=diamonds, 
            scatter=False
           )
```

```python
# Changing the range of axis

plt.gca().set(xlim=(0, 2))
plt.gca().set(ylim=(0, 12500))

sns.regplot(x='carat', y='price', data=diamonds, scatter=False)
```

# regplot Options

# Confidence Intervals

```python
sns.regplot(x='carat', y='price', data=diamonds,
           ci = None)
```

```python
# Customize a confidence interval 
# ci=95 shows 95% confidence intervals

sns.regplot(x='carat', y='price', data=diamonds,
           ci = 95) 
```

# Jitter
Jitter is useful when you work with a **discrete variable** such as the number of children in a family, type of fruit, etc. You can't have a value between two discrete values. For example, you can't have 2.5 children.

```python
mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
sns.regplot(x="cylinders", y="weight", data=mpg)
```

```python
sns.regplot(x="cylinders", y="weight", data=mpg,
            x_jitter=.15,
            scatter_kws={'alpha': 0.2}
           )
```

# Models
If the data points don't align well with a linear regression line, you might consider a more complex model with a curve.

Seaborn offers four types of non-linear models:
1. Polynomial Regression (order):
   
   Adjusts the curve's shape using a polynomial equation.
   
3. Robust Regression:
   
   Handles outliers and provides more reliable estimates.

   
5. Logistic Regression:
   
   models binary outcomes (e.g., yes/no, true/false).

   
7. Lowess (Locally Weighted Scatterplot Smoothing):
   
   Fits a smooth curve to the data, especially useful for noisy data.

# Polynomial Regression (order)
Polynomial Regression is a statistical model that fits a nonlinear relationship between a dependent variable and one or more independent variables. Unlike linear regression, which assumes a linear relationship, polynomial regression allows for more complex curves to be modeled. 

Use `Order= ` to capture nonlinear trends.


Order in polynomial regression refers to the highest power of the independent variable in the polynomial equation.

Order 1 = linear regression

Order 2 = quadratic regression

Order 3 = cubic regression

```python
sns.regplot(x='carat', y='price', data=diamonds,
            fit_reg=False
               )
```

```python
sns.regplot(x='carat', y='price', data=diamonds,
            order=2
           )
```

# Robust Regression
Robust Regression is a statistical method designed to handle outliers or data points that deviate significantly from the majority of the data. Unlike traditional least squares regression, which can be heavily influenced by outliers, robust regression techniques are less sensitive to such points.

Use `robust=True` to assign less weight to outliers.

```python
x_example=[0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]
y_example=[0.1, 0.8, 2.2, 2.7, 3.8, 4.5, 6.2, 6.8, 7.9, 9.4, 30.4]
```

```python
sns.regplot(x=x_example, 
            y=y_example,
            ci=None
            )
```

```python
sns.regplot(x=x_example, 
            y=y_example,
            ci=None,
            robust=True
            )
```
